```python
# Set input data and target data
X = data[["Attended"]].to_numpy()
Y = data[["Will take exam"]].to_numpy()

# creare clusters and call the .fit() method
fcm = FCM(n_clusters = 3)
fcm.fit(X)

# Set lables and centers
fcm_centers = FCM.centers
fcm_labels = FCM.predict
```

```python
# Number of samples in the data
n_samples = 7

X = np.concatenate((
    np.random.normal((-2, -2), size=(n_samples, 2)),
    np.random.normal((2, 2), size=(n_samples, 2))))
```

```python
# fit FCM
fcm = FCM(n_clusters=2)
fcm.fit(X)
```

```python
# outputs
fcm_centers = fcm.centers
fcm_labels = fcm.predict(X)

# plot result
f, axes = plt.subplots(1, 2, figsize=(15,7))
axes[0].scatter(X[:,0], X[:,1], alpha=.1)
axes[1].scatter(X[:,0], X[:,1], c = fcm_labels, alpha=.1)

axes[1].scatter(fcm_centers[:,0], fcm_centers[:,1], marker="*", s=500, c="g")

plt.show()
```
